### 4. Make a request to the wikipedia page

Retrieve the whole html marks of the page and split the string every time the <strong>p</strong> tag appeared. This was done to get only the "html" of the table. Then I store in the tableCanada variable the html <<-table-> string and finally transformed into a DataFrame

```python
listCanada = wikipedia.page('List_of_postal_codes_of_Canada:_M')
print(type(listCanada.html()))
htmlCanada = listCanada.html()
arrayCanada = htmlCanada.split('</p>\n')
tableCanada = arrayCanada[2].split('\n<h2>')
tableCanada = tableCanada[0]

df_Canada = pd.read_html(tableCanada)
```

```python
df_Canada = df_Canada[0]
df_Canada
```
